```python
#install qiskit module to the notebook and import libraries and methods
!pip install qiskit --quiet
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit

#define a quantum reg using the QuantumRegister object
qreg = QuantumRegister(2)
```

```python
#you can call it by any valid name
output = QuantumRegister(1)
```

```python
#define a classical register using the ClassicalRegsister object
creg = ClassicalRegister(4)
```

```python
#define a quantum circ using the QuantumCircuit object
circuit = QuantumCircuit(qreg, creg)
```

```python
#you can include multiple quantum registers in a QuantumCircuit
circuit2 = QuantumCircuit(qreg, creg, output)
```

```python
#you can omit the classical reg if no measurement is taken at the end of the circuit
circuit3 = QuantumCircuit(qreg, output)
```

```python
#you can also define a circuit without defining the registers
circuit1 = QuantumCircuit(2)
```
